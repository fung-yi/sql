```python
## finding duplicate row

sql="""
        with RowNumCTE as(
        select *,
         row_number() over (
         partition by ParcelID,
             PropertyAddress,
             SalePrice,
             SaleDate,
             LegalReference
             ORDER BY
             UniqueID
             ) as row_num
        from NashvilleHousing
        )
        select *
        from RowNumCTE
        where row_num > 1
        order by PropertyAddress

    """
df = pd.read_sql_query(sql, conn)
df
```

```python
## delete duplicate row

sql="""
        with RowNumCTE as(
        select *,
         row_number() over (
         partition by ParcelID,
             PropertyAddress,
             SalePrice,
             SaleDate,
             LegalReference
             ORDER BY
             UniqueID
             ) as row_num
        from NashvilleHousing
        )
        delete
        from RowNumCTE
        where row_num > 1
    """

cursor.execute(sql)

## check if all the duplicate rows are removed
sql="""
        with RowNumCTE as(
        select *,
         row_number() over (
         partition by ParcelID,
             PropertyAddress,
             SalePrice,
             SaleDate,
             LegalReference
             ORDER BY
             UniqueID
             ) as row_num
        from NashvilleHousing
        )
        select *
        from RowNumCTE
        where row_num > 1
        order by PropertyAddress

    """
df = pd.read_sql_query(sql, conn)
df
```
